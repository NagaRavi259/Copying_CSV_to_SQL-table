## Checking the database data insertions using the Python and SQLalchemy
- in this file we are going the write a script to check the database for any data missing using python

```python
from sqlalchemy import create_engine
import sqlalchemy
import pandas as pd
from urllib.parse import quote  
```

```python
##Connecting dataBase using the python and sqlalchemy and python
engine = create_engine('postgresql://postgres:%s@localhost:5432/postgres' % quote('postgres'))
```

```python
## reading Sql data using the pandas and storing into a varibale
df = pd.read_sql_table("Table_Name", engine)
```

```python
## checking the Dataframe Cratedtime column data type
type(df["CreatedTime"][1])
```

```python
df
```

```python
## Sorting the column to simplify the process
df = df.sort_values('CreatedTime', ascending=False)
```

```python
## Checking the time differences between to every two rows
df['CreatedTime'].diff(-1)>pd.to_timedelta('2s')
```

```python
## Pandas way to Checking the time differences between to every two rows
df['CreatedTime'].diff(-1)
```

```python
## getting the all rows with missing entries
df_2=df.loc[df['CreatedTime'].diff(-1)>pd.to_timedelta('2s')]
```
